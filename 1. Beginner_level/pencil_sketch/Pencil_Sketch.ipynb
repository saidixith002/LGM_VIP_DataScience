## Procedure:
#### Step 1: Reading Image
#### Step 2: Convert to Gray Image
#### Step 3: Invert the Gray Image
#### Step 4: Blur the Inverted Image
#### Step 5: Pencil Sketch
___________

```python
import cv2
from matplotlib import pyplot as plt
```

### Step:1 Reading Image

```python
img=cv2.imread("bird.jpg")
```

```python
plt.imshow(img)
plt.title("Original Image")
plt.show()
```

### Step 2: Convert to Gray Image

```python
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray , cmap="gray")
plt.title("Gray Image")
plt.show()
```

### Step 3: Invert the Gray Image

```python
inverted_gray_image = 255 - gray
plt.imshow(inverted_gray_image,cmap="gray")
plt.title("Inverted Image")
plt.show()
```

### Step 4: Blur the Inverted Image

```python
blurred_image = cv2.GaussianBlur(inverted_gray_image,(21,21),0)
inverted_gray_image=255 - gray
plt.imshow(blurred_image,cmap="gray")
plt.title("Blurred Image")
plt.show()
```

#### Step 5: Pencil Sketch

```python
pencil_sketch = cv2.divide(gray, inverted_gray_image, scale=256.0)
plt.imshow(pencil_sketch,cmap="gray")
plt.title("Pencil Sketch")
```
